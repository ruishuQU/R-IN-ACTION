```r
# 5-1 均值和标准差的计算
x <- c(1,2,3,4,5,6,7,8)
mean(x)#简洁写法
sd(x)
```

```r
n <- length(x)
meanx <- sum(x)/n
css <- sum((x - meanx)^2)
sdx <- sqrt(css / (n-1))
meanx# 复杂写法 
sdx
```
